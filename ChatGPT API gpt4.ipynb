```python
i=0
# Loop through your questions and get answers for each one


start_time = time.time()

for question in questions1000["ques"]:
    i=i+1    
    print(i)   
    if(i%80==0):
        time.sleep(61)
    try:        
        output = openai.ChatCompletion.create(
        model="gpt-4", 
        messages=[{"role": "user", "content": question
                               }]
        )
        #df.append(output['choices'][0]['message']['content'])
        df=df.append({'Srno': i, 'Answer': output['choices'][0]['message']['content']}, ignore_index=True)

    except Exception as e:
        #answers.append("An error occurred:"+ str(e))
        df=df.append({'Srno': i, 'Answer': answers.append("An error occurred:"+ str(e))}, ignore_index=True)
    # Add your answers to the DataFrame

end_time = time.time()
# Calculate the elapsed time
elapsed_time = end_time - start_time    

questions1000["Answer"] = df['Answer']
#questions["Answer"] = df['Answer']

# Print out the whole output dictionary
# Get the output text only
```
